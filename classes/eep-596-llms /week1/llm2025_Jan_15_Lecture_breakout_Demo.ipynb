```python
# prompt: numpy two layer hard coded weights perceptron that calculates xor and plots the intermeter weights

import numpy as np
import matplotlib.pyplot as plt
####################
#EDIT THESE VARIABLES

# Initialize weights for the two layers
# Hardcoded weights for demonstration
W1 = np.array([[1, -1], [-1, 1]])
b1 = np.array([-0.5, -0.5])
W2 = np.array([1, 1])
b2 = np.array([-0.5])

#####################

# Input data for XOR
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])

# Define the activation function (step function)
def activation(x):
  # return 1 if x >= 0 else 0 # step function
  return x if x >= 0 else 0


# Store intermediate weights for plotting
intermediate_weights = []


# Perform forward pass for each input
for i in range(len(X)):
    # Layer 1
    z1 = np.dot(X[i], W1.T) + b1
    a1 = np.array([activation(z1[0]), activation(z1[1])])

    intermediate_weights.append(a1)

    # Layer 2
    z2 = np.dot(a1, W2) + b2
    a2 = (z2)

    print(f"Input: {X[i]}, Output Prestep: {a2}, Output Poststep: {1 if a2 > 0 else 0}, Expected: {y[i]}")


# Plot intermediate weights
plt.figure(figsize=(8, 6))
for i in range(len(intermediate_weights)):
    plt.scatter(intermediate_weights[i][0], intermediate_weights[i][1], label=f'Input {X[i]}', marker="oxox"[i], s=100)
plt.xlabel("Neuron 1 Output")
plt.ylabel("Neuron 2 Output")
plt.title("Intermediate Weights of a Two-Layer Perceptron for XOR")
plt.xlim(-0.5, 1.5) #adjust limits to better see the plots
plt.ylim(-0.5, 1.5) #adjust limits to better see the plots
plt.legend()
plt.grid(True)
plt.show()
```
