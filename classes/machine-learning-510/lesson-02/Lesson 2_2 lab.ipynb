## <h1>Linear Regression</h1>

<B>Let's generate some data that shows roughly linear relationship between page speed and amount purchased</B>

```python
%matplotlib inline
import numpy as np
from pylab import *

pageSpeeds = np.random.normal(3.0, 1.0, 1000)
print(pageSpeeds)
```

```python
purchaseAmount = 100 - (pageSpeeds + np.random.normal(0, 0.1, 1000)) * 3

scatter(pageSpeeds, purchaseAmount)
```

<b>As we only have two features, we can keep it simple and just use scipy.state.linregress:</b>

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchaseAmount)
```

```python
import matplotlib.pyplot as plt

def predict(x):
    return slope * x + intercept

fitLine = predict(pageSpeeds)

plt.scatter(pageSpeeds, purchaseAmount)
plt.plot(pageSpeeds, fitLine, c='r')
plt.show()
```

<b>Not surprisingly, our r-squared value shows a really good fit</b>
